```python
import pandas as pd
import numpy as np
from random import choice, randint
import random
from faker import Faker

# Inisialisasi Faker
fake = Faker('id_ID')

# Daftar lokasi area
locations = {
    "Main Production Area": ["Extruder A", "Injection Molder B", "Blower C", "Thermoformer D", "Molder E", "PPIC", "QA","Mekanik", "Utility", "Elektrik", "R&D"],
    "Extrusion": ["Extruder A"],
    "Injection Molding": ["Injection Molder B"],
    "Blow Molding": ["Blower C"],
    "Thermoforming": ["Thermoformer D"],
    "Metalizing": ["Mekanik", "Utility", "Elektrik", "Utility", "Metalize"],
    "Feeding": ["Extruder A", "Blower C"],
    "Washing Area": ["Mekanik", "Utility", "Elektrik", "Utility"],
    "Quality Control Area": ["QA"],
    "Raw Material Storage": ["Warehouse", "Logistics"],
    "Finished Goods Storage": ["Warehouse", "Logistics"],
    "Support Material Storage": ["Warehouse", "Logistics"],
    "Hazardous Material Warehouse": ["Warehouse", "Logistics"],
    "Slitting" : ["SLitter", "Mekanik", "Utility", "Elektrik"],
    "Canteen Area": ["HR", "R&D", "Finance", "IT"],
    "Prayer Room": ["HR", "R&D", "Finance", "IT"],
    "Break Room": ["HR", "R&D", "Finance", "IT"],
    "Office Area": ["Marketing", "Finance", "HR", "IT"]
}

# Daftar departemen
departments = ["Extruder A", "Injection Molder B", "Blower C", "Thermoformer D", "Molder E",
               "PPIC", "QA", "R&D", "HR", "Marketing", "Finance", "IT",
               "Logistics", "Warehouse", "Mekanik", "Utility", "Elektrik", "Metalize"]

# Daftar nama jalan di Surabaya, Sidoarjo, Gresik, dan sekitarnya
roads = [
    "Jalan Raya Darmo, Surabaya", "Jalan Basuki Rahmat, Surabaya", "Jalan Diponegoro, Sidoarjo",
    "Jalan Gubernur Suryo, Surabaya", "Jalan Raya Taman Pinang, Sidoarjo", "Jalan Veteran, Gresik",
    "Jalan Ahmad Yani, Surabaya", "Jalan Kartini, Sidoarjo", "Jalan Sunan Giri, Gresik"
]

# Daftar jenis bahaya
hazards = ["Fisik", "Kimia", "Listrik", "Mekanik"]

# Daftar penyebab insiden
causes = ["Faktor Manusia", "Alat Rusak", "Kesalahan Prosedur", "Lingkungan Kerja"]

# Daftar kondisi korban detail sesuai dengan jenis insiden
detailed_conditions = {
    "First Aid": ["Luka Bakar Ringan", "Tergores", "Terkena Benturan Ringan", "Tersengat Listrik Ringan"],
    "Medical Treatment": ["Luka Jahitan", "Cedera Ringan", "Luka Robek", "Luka Dalam"],
    "Restricted": ["Patah Tulang (Tangan)", "Cedera Otot", "Cedera Lutut", "Cedera Pergelangan"],
    "Fatality": ["Cedera Fatal", "Kematian"]
}

# Fungsi untuk menentukan waktu insiden dengan distribusi 30-40% pada jam malam
def random_time():
    if random.random() < 0.7:  # 70% chance for incidents to occur between 10 PM - 7 AM
        # Determine hour range: 10 PM - 11 PM (22-23) or midnight to 6 AM (0-6), or 7 AM (7)
        hour = randint(22, 23) if random.random() < (2/9) else randint(0, 6)
    else:  # 30% chance for incidents during regular working hours (8 AM - 9 PM)
        hour = randint(8, 21)
    minute = randint(0, 59)
    return f"{hour:02}:{minute:02}"

# Membuat DataFrame
data = []
for i in range(100):
    # Tentukan jenis kelamin secara acak
    gender = choice(["Laki-laki", "Perempuan"])

    # Tentukan nama berdasarkan jenis kelamin
    if gender == "Laki-laki":
        name = fake.first_name_male() + " " + fake.last_name()
    else:
        name = fake.first_name_female() + " " + fake.last_name()

    age = randint(20, 48)
    department = choice(departments)

    # Pilih lokasi yang sesuai dengan departemen
    possible_locations = [loc for loc, depts in locations.items() if department in depts]
    location = choice(possible_locations)

    hazard = choice(hazards)
    cause = choice(causes)
    lti_duration = randint(0, 30)  # Durasi LTI dalam hari
    damage_cost = randint(1000000, 50000000)  # Biaya kerusakan
    medical_cost = randint(100000, 5000000)  # Biaya pengobatan

    # Menentukan jenis insiden berdasarkan proporsi
    rand_num = random.random()
    if rand_num < 0.15:
        incident_type = "First Aid"
    elif rand_num < 0.80:
        incident_type = "Medical Treatment"
    elif rand_num < 0.95:
        incident_type = "Restricted"
    else:
        incident_type = "Fatality"

    # Menentukan jika ini adalah kecelakaan lalu lintas
    if random.random() < 0.1:  # 10% kemungkinan kecelakaan lalu lintas
        incident_type = choice(["Medical Treatment", "Restricted", "Fatality"])
        is_traffic = "Ya"
        location = choice(roads)
        condition = "Cedera Akibat Kecelakaan Lalu Lintas"
    else:
        is_traffic = "Tidak"
        condition = choice(detailed_conditions[incident_type])

    # Menambahkan data ke list
    data.append([i + 1, name, gender, age, pd.Timestamp(f"2023-{randint(1,12):02}-{randint(1,28):02}"),
                  random_time(), incident_type, location, department, hazard, cause, condition,
                  lti_duration, damage_cost, medical_cost, is_traffic])

# Membuat DataFrame
columns = ["Nomor", "Nama_Korban", "Jenis_Kelamin", "Usia", "Tanggal_Insiden", "Waktu_Insiden",
           "Jenis_Insiden", "Lokasi_Insiden", "Departemen_Terkait", "Jenis_Bahaya", "Penyebab_Insiden",
           "Kondisi_Korban", "Durasi_LTI", "Biaya_Kerusakan", "Biaya_Pengobatan", "Kecelakaan_Lalu_Lintas"]

df = pd.DataFrame(data, columns=columns)

# Menampilkan 5 baris pertama
df.head()
```
